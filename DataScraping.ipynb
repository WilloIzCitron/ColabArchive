```python
from bs4 import BeautifulSoup
import requests
import pandas

web = "https://genshin-impact.fandom.com/wiki/Character/List"
pg = requests.get(web)
res = BeautifulSoup(pg.content, "html.parser")
data = []
rows = res.find("tbody").find_all("tr")
for row in rows:
    cols = row.find_all(["td", "th"])
    cols = [col.text.strip() for col in cols]
    if cols:
        data.append(cols)


pd = pandas.DataFrame(data)
pd = pd.iloc[1:].reset_index(drop=True)
pd
```

```python
from matplotlib import pyplot as plt
import seaborn as sns
pd.groupby(8).size().plot(kind='barh', color=sns.palettes.mpl_palette('Dark2'))
plt.gca().spines[['top', 'right',]].set_visible(False)
```
