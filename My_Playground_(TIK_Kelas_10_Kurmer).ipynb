# kita coba array ya (Ranai)

```python
# array adalah OBJEK!!! suatu deret data yang berisi beberapa data.
data = [10, 9, 7, 8, 10, 8]
print(data[1])
data[2] = 10
print(data)
data.append(9)
print(data)
data.pop(2)
print(data)
```

fungsi append() pada array itu berfungsi sebagai memasukan data kedalam array tersebut, dan kegunaan data[index] sebagai selector yang dimana menunjukan ke indeks yang dituju seperti indek ke-1 pada array data maka hasilnya 9. jika indeks melebihi jumlah data (data.length() + 1) maka menghasilkan IndexError.

```python
data = [10, 9, 7, 8, 10 ,8]
indeks = 0
for elemen in data:
  print("Elemen ke ", indeks, " = ", elemen)
  indeks = indeks + 1
```

# While loop

```python

datat = 1
while True:
  print(datat)
  datat = datat + 1
  if datat >= 11:
    break

# while adalah looping yang berulang dan tidak terhingga. untuk menghentikannya, gunakan break
```

# try n' except

```python
data = [10, 9, "Cindi", 4, "8"]
for x in data:
    print(x/2)
```

```python
data = [10, 9, "Cindi", 4, "8"]
for x in data:
  try:
    print(x/2)
  except:
    print("Bukan bilangan")
```

Untuk mengambil satu elemen HTML, gunakan find() di library BeautifulSoup. seperti dibawah ini:

```python
import requests
from bs4 import BeautifulSoup
from PIL import Image

website = "https://genshin-impact.fandom.com/wiki/Citlali"
page = requests.get(website)
res = BeautifulSoup(page.content, "html.parser")
image = res.find('a', attrs={"class": "image image-thumbnail", "title": "Card"}).get("href")
with open("/content/cache.png", 'wb') as handler:
   handler.write(requests.get(image).content)
pic = Image.open("/content/cache.png")
width, height = pic.size
mode = pic.mode
pic
```
